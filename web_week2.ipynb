```python
# 不会处理爬虫出现缺失值的问题，公司链接未完成
from requests_html import HTMLSession
import pandas as pd
session = HTMLSession()
payload = {'key':'web',
          'dqs':'050020',
           'curPage':'3'
          }
headers = {
    'User-Agent':"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/13.0.5 Safari/605.1.15"
}
r = session.get("https://www.liepin.com/zhaopin",headers=headers,params =payload )
news = r.html.xpath('//div[@class="job-info"]/h3/a')
news2 = r.html.xpath('//p[@class="company-name"]/a/text()')
new3 = r.html.xpath('//div[@class="job-info"]/p/span[@class="text-warning"]/text()')
new4 = r.html.xpath('//*[@class="area"]/text()')
new5 = r.html.xpath('//div[@class="job-info"]/p/span[@class="edu"]/text()')
new6 = r.html.xpath('//div[@class="job-info"]/p[@class="condition clearfix"]//@title')
new8 = r.html.xpath('//*[@class="time-info clearfix"]/time/text()')
new9 = r.html.xpath('//*[@class="job-info"]/h3/a//@href')
#new11 = r.html.xpath('//*[@class="company-name"]/a//@href')
new7 = []
new10 = []
news1 = []
for u in news:
    news1.append(u.text)
for i in new6:
    new7.append(i.split( '_')[3])
for n in new9:
    if 'https' not in n:
        new10.append(('{}'.format("https://www.liepin.com")+n))
    else:
        new10.append(n)
results = pd.DataFrame ( {
        "职位": news1,
        "公司名称": news2,
        "薪水": new3,
        "地点": new4,
        "学历要求":new5,
        "经验":new7,
        "时间":new8,
        "链接":new10,
        #"公司链接":new11,
    
      } )
display(results)
results.to_csv("liepin_test.csv", encoding='utf_8_sig')
```
